```
from pyspark.sql.functions import sum
df1.select("Customer_ID","Buy_Times").groupBy("Customer_Id").agg(sum("Buy_Times").alias("sum_buy_times")).show(5)
```

```
from pyspark.sql.types import IntegerType
df2 = df1.withColumn("Buy_Times",df1["Buy_Times"].cast(IntegerType()))
```

```
df2.groupBy("Customer_Id").agg(sum("Buy_Times"),max("Buy_Times")).show(5)
```

```
from pyspark.sql.functions import isnan, when, count, col
import pyspark.sql.functions as F
df2.select([count(when(isnan(c) | col(c).isNull(), c)).alias(c) for c in df2.columns]).show()
```
